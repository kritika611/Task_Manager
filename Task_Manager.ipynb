```python
import datetime
def user_information(ussnm, pssd):
    name = input("enter your name please: ")
    address = input("your address")
    user_id = input("Your id please")
    ussnm_ = ussnm+"_"+"task.txt"
    print(ussnm_)
     
    f = open(ussnm_, 'a')
    f.write(pssd)
    f.write("\nName: ")
    f.write(name)
    f.write('\n')
    f.write("Address :")
 
    f.write(address)
    f.write('\n')
    f.write("user_id :")
    f.write(user_id)
    f.write('\n')
    f.close()
    
import hashlib

def hash_password(password):
   password_bytes = password.encode('utf-8')
   hash_object = hashlib.sha256(password_bytes)
   return hash_object.hexdigest()

import hashlib
def signup():
    print("Please enter the username by which you \
    wanna access your account")
    username = input("please enter here:  ")
    password =  hash_password(input("Enter a password:  "))
    user_information(username, password)
    print("Registration Successful")
    login()
    
def login():
    print("Please enter your username ")
    user_nm = input("Enter here: ")
     
    # Password as entered while logging in
    #pssd_wr = (input("enter the password:"))+'\n'
    pssd_wr = hash_password(input("Enter a password:  "))+'\n'
    usernm = user_nm+"_"+"task.txt"
    print(usernm)
    f_ = open(usernm, 'r')
    k = f_.readlines(0)[0]
    f_.close()
         
         # Checking if the Password entered is same as
         # the password saved while signing in
    if pssd_wr == k: 
        print(
                "1--to view your data \n2--To add task \n3--mark as completed \n4task delete \n5--exit")
        a = input()
        if a == '1':
            view_data(usernm)
        elif a == '2':
            add_task(usernm)
        elif a == '3':
            task_completed(usernm)
        elif a == '4':
            delete_task() 
        elif a == '5':
            print("logout")
            exit()
        else:
            print("Wrong input ! ")
    else:
        print("YOUR PASSWORD OR USERNAME IS WRONG , Plz enter Again")
        login()
        
def view_data(username):
    user_nm = input("Enter here: ")
    usernm = user_nm+"_"+"task.txt"
    ff = open(usernm, 'r')
    print(ff.read())
    ff.close()

def add_task(username):
    user_nm = input("Enter here: ")
    usernm = user_nm+"_"+"task.txt"
    print("enter n.o of task you want to ADD")
    j = int(input())
    f1 = open(username, 'a')
     
    for i in range(1, j+1):
        task = input("enter the task")
        target = input("enter the target")
        pp = "TASK "+str(i)+' :'
        qq = "TARGET "+str(i)+" :"
         
        f1.write(pp)
        f1.write(task)
        f1.write('\n')
        f1.write(qq)
        f1.write(target)
        f1.write('\n')
         
        print("Do u want to stop press space bar otherwise enter")
        s = input()
        if s == ' ':
            break
    f1.close()
    
def task_completed(username):
    user_nm = input("Enter here: ")
    usernm = user_nm+"_"+"task.txt"
    print("Please enter the tasks ID which are completed ")
     
    task_completed = input()
    fw = open(usernm, 'a')
    DT = str(datetime.datetime.now())
     
    fw.write(DT)
    fw.write("\n")
    fw.write("COMPLETED TASK \n")
    fw.write(task_completed)
    fw.write("\n")
    print("Task Completed")
    
def delete_task():
    user_nm = input("Enter username here: ")
    file_name = user_nm+"_"+"task.txt"
    user = input("Enter the task id you want to delete: ")
    task_number = int(user)
    #fw = open(username, 'a')
    with open(file_name) as f:
        tasks = f.readlines()
        if 1 <= task_number <= len(tasks):
            tasks.pop(task_number - 1)
            with open(file_name, "w") as f:
                f.writelines(tasks)
                print("Task deleted successfully!")
        else:
            print("Invalid task number!")
            
def task_manager():
    
    print("WELCOME TO Kritika`S TASK MANAGER")
    print("are you new to this software")
    a = int(input("Type 1 if new otherwise press 0 ::"))
   
    if a == 1:
        signup()
    elif a == 0:
        login()
    else:
        print("You have provided wrong input !")        
```
